```python
import numpy as np
import pandas as pd

excel_file = pd.ExcelFile('study.xlsx')

# 获取指定工作表中的数据
df = excel_file.parse('Sheet1')

df.head()
```

```python
# 查看每列缺失值数量
missing_values = df.isnull().sum()
print("缺失值统计：\n", missing_values)
```

```python
# 删除包含缺失值的行
df_dropna = df.dropna()

# 删除全部为缺失值的列
df_dropna_col = df.dropna(axis=1, how='all')
```

```python
# 提取数值型列
numeric_columns = df.select_dtypes(include=[np.number]).columns

# 计算每个数值型列的均值和标准差
means = df[numeric_columns].mean()
stds = df[numeric_columns].std()

# 定义 Z-Score 阈值
z_score_threshold = 3

# 识别异常值
outliers = {}
for col in numeric_columns:
    z_scores = np.abs((df[col] - means[col]) / stds[col])
    outliers[col] = df[z_scores > z_score_threshold][col]

# 输出异常值情况
for col, outlier_values in outliers.items():
    if not outlier_values.empty:
        print(f'列 {col} 中的异常值：')
        print(outlier_values)

# 将异常值替换为 NaN
for col in numeric_columns:
    z_scores = np.abs((df[col] - means[col]) / stds[col])
    df[col] = np.where(z_scores > z_score_threshold, np.nan, df[col])
```
